```python
import pandas as pd
import numpy as np

# Creamos el DataFrame manualmente
data = {
    "id": list(range(1, 11)),                          # 1 a 10 (autoincremental)
    "documentacion_completa": ["Sí", "No", "Sí", None, "No", "Sí", "No", "Sí", "No", "Sí"],  # dato faltante en fila 4
    "nota": [5, 3, 9, "ocho", 1, 7, 10, np.nan, 6, 2],  # inconsistencia en fila 4 (texto en lugar de número)
    "turno": ["mañana", "tarde", "mañana", "mañan", "tarde", "mañana", None, "tarde", "mañana", "tarde"]
    # en fila 4 escribimos "mañan" (error tipográfico) y fila 7 tiene dato faltante
}

df = pd.DataFrame(data)
df.to_csv('archivo.csv', index=False)
print(df)
```

```python
#objeto alumno y metedos de comprobación
class Alumno:
    def __init__(self, id, documentacion_completa, nota, turno):
        self.id = id
        self.documentacion_completa = documentacion_completa
        self.nota = nota
        self.turno = turno

    def es_documentacion_completa(self):
        """Debe responder 'Sí' para estar completo"""
        return self.documentacion_completa == "Sí"

    def aprobo_el_ingreso(self):
      try:
        return int(self.nota) >= 6
      except (ValueError, TypeError):
        return False

    def tiene_turno_valido(self):
        """Debe ser 'mañana' o 'tarde'"""
        return self.turno in ["mañana", "tarde"]

    def es_aspirante(self):
        """Cumple todos los requisitos"""
        return (self.es_documentacion_completa() and
                self.aprobo_el_ingreso() and
                self.tiene_turno_valido())

    def __repr__(self):#un método especial de Python que define cómo se muestra un objeto cuando lo imprimís o cuando aparece en una lista o en el intérprete.
        return f"Alumno({self.id}, {self.documentacion_completa}, {self.nota}, {self.turno})"

```

```python
#  Convertimos cada fila en un objeto Alumno
alumnos = [Alumno(row.id, row.documentacion_completa, row.nota, row.turno) for row in df.itertuples(index=False)]

#  Verificamos si cada alumno cumple con los requisitos
for alumno in alumnos:
    print(f"{alumno} → Aspirante válido: {alumno.es_aspirante()}")
```

```python
df["es_documentacion_completa"] = df.apply(
    lambda fila: Alumno(
        fila["id"],
        fila["documentacion_completa"],
        fila["nota"],
        fila["turno"]
    ).es_documentacion_completa(),
    axis=1
)

df["aprobo_el_ingreso"] = df.apply(
    lambda fila: Alumno(
        fila["id"],
        fila["documentacion_completa"],
        fila["nota"],
        fila["turno"]
    ).aprobo_el_ingreso(),
    axis=1
)

df["turno_valido"] = df.apply(
    lambda fila: Alumno(
        fila["id"],
        fila["documentacion_completa"],
        fila["nota"],
        fila["turno"]
    ).tiene_turno_valido(),
    axis=1
)

df["admitido"] = df.apply(
    lambda fila: Alumno(
        fila["id"],
        fila["documentacion_completa"],
        fila["nota"],
        fila["turno"]
    ).es_aspirante(),
    axis=1
)

print(df)
```

# Resultados
Sea crean los archivos admitidos.cvs y no_admitidos.csv

```python
df_admitidos = df[df["admitido"]]
df_admitidos.to_csv('admitidos.csv', index=False)

df_no_admitidos = df[~df["admitido"]]
df_no_admitidos.to_csv('no_admitidos.csv', index=False)
```
